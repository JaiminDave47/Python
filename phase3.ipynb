# Practice with real world data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000]  # Chai Point
])

print("==== Zomato sales analysis ====")
print("\nSales data shape:", sales_data.shape)
print("\nSample data for first 3 restaurants:\n", sales_data[:3,:])
```

## Yearly Sales (From every restro)

```python
print("Total Sales per year: ")
# for i in range(0, 4):
#     print(f"Year {2021+i}: ", np.sum(sales_data[:, i+1]))
print(np.sum(sales_data[:, 1:], axis=0))
```

## Min/Max sales per restaurant

```python
print("Minimum sale: ", np.min(sales_data[:,1:], axis=1))
print("Maximum sale: ", np.max(sales_data[:,1:], axis=1))
```

## Min/Max sales per year

```python
print("Minimum sale (Yearly): ", np.min(sales_data[:,1:], axis=0))
print("Maximum sale (Yearly): ", np.max(sales_data[:,1:], axis=0))
```

## Average sales per restaurant

```python
print("Average sales (Restaurant wise):", np.mean(sales_data[:, 1:], axis=1))
```

## Cumulative sales

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative sales (Restaurant wise):\n", cumsum)
```

## Ploting the graph

```python
plt.figure(figsize=(10, 5))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative sales across all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

# Vectors

```python
vec1 = np.array([1, 2, 3])
vec2 = np.array([4, 5, 6])
print("\nSum of vec1 and vec2:", vec1 + vec2)
print("\nMultiplication:", vec1 * vec2)
print("\nDot Product:", np.dot(vec1, vec2))

angle = np.arccos(np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2)))
print("\nAngle: ", angle)
```

## Vectorized Operations

```python
restaurant_types = np.array(['biryani', 'pizza', 'chinese', 'burger'])
vectorized_upper = np.vectorize(str.upper)
print("Vectorized Upper: ", vectorized_upper(restaurant_types))
```
